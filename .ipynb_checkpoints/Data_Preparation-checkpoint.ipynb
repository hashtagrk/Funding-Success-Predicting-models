## Get Location

```python
def get_location(state_code):
    
    statename_mappng={'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona',
                      'AR': 'Arkansas', 'CA': 'California', 'CO': 'Colorado', 
                      'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 
                      'GA': 'Georgia', 'HI': 'Hawaii', 'ID': 'Idaho', 
                      'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 
                      'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 
                      'ME': 'Maine', 'MD': 'Maryland', 'MA': 'Massachusetts', 
                      'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 
                      'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 
                      'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey', 
                      'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina', 
                      'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma', 
                      'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 
                      'SC': 'South Carolina', 'SD': 'South Dakota', 'TN': 'Tennessee', 
                      'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont', 'VA': 'Virginia', 
                      'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming','DC':'DC','AE':'AE'}
    NE=['Maine', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New Hampshire', 'Vermont', 
        'New York', 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland']
    SE=['West Virginia', 'Virginia', 'Kentucky', 'Tennessee', 'North Carolina', 'South Carolina', 'Georgia, Alabama', 
        'Mississippi', 'Arkansas', 'Louisiana', 'Florida']
    MW=['Ohio', 'Indiana', 'Michigan', 'Illinois', 'Missouri', 'Wisconsin', 'Minnesota', 'Iowa', 'Kansas', 'Nebraska', 'South Dakota', 
       'North Dakota'] 
    SW=['Texas', 'Oklahoma', 'New Mexico', 'Arizona']
    W=['Colorado', 'Wyoming', 'Montana', 'Idaho', 'Washington', 'Oregon', 'Utah', 'Nevada', 'California','Alaska', 'Hawaii','DC']
    
    if statename_mappng[state_code] in NE:
        return 'northeast'
    elif statename_mappng[state_code] in SE:
        return 'southeast'
    elif statename_mappng[state_code] in MW:
        return 'midwest'
    elif statename_mappng[state_code] in SW:
        return 'southwest'
    elif statename_mappng[state_code] in W:
        return 'west'
    else:
        return 'others'
    
```

## Get Age of Company in Months

```python
from datetime import datetime
from datetime import date
from dateutil import relativedelta
def months_between(d1):
    if d1=='':
        return 0
    date1 = datetime.strptime(str(d1), '%Y-%m-%d')
    date2 = datetime.strptime(str(datetime.today()).split()[0], '%Y-%m-%d')
    return (date2.year - date1.year) * 12 + date2.month - date1.month
```

```python
orgs_final['age_months'] = orgs_final['founded_on'].apply(lambda x:months_between(x))
```

```python
import re
def ret_binned_degree(x):
    x=x.strip(')')
    x=x.strip('(')
    if 'Dr' in x or re.search('phd', x, re.IGNORECASE) or re.search('ph.d', x, re.IGNORECASE) or re.search('doctorate', x, re.IGNORECASE):
        return 'PHD'
    elif re.search('mtech', x, re.IGNORECASE) or re.search('mba', x, re.IGNORECASE) or re.search('master', x, re.IGNORECASE) or re.search('m.tech', x, re.IGNORECASE) or re.search('msc', x, re.IGNORECASE) or re.search('m.sc', x, re.IGNORECASE):
        return  'Master'
    elif re.search('bach', x, re.IGNORECASE) or x.startswith('B') or re.search('bsc', x, re.IGNORECASE):
        return 'Bachelor'
    elif re.search('dipl', x, re.IGNORECASE) or 'Dip' in x:
        return 'diploma'
    elif re.search('assoc', x, re.IGNORECASE) or re.search('AA', x, re.IGNORECASE) :
        return 'associate'
    else:
        return 'others'
```

```python
degree_data['d_tag'] = degree_data['degree_type'].apply(lambda x:ret_binned_degree(x))
```

```python
import operator
import math
import collections
def ret_final_tag(x):
    tag_dic_pref={'PHD':5, 'Bachelor':3,'Master':4 ,'diploma':2,'associate':1,'others':0}
    tag_dic_pref_inverse={5:'PHD', 3:'Bachelor',4:'Master' ,2:'Diploma',1:'associate',0:'others'}
    if x=='':
        return 'others'
    l1=x.split(',')
    if 'others' in l1:
        l1.remove('others')
    if len(l1)==0:
        return 'others'
    
    l=[]
    ini=len(l1)
    for itm in l1:
        itm=itm.strip()
        l.append(itm)
    
    freq = {}
    for item in l:
        freq[item] = freq.get(item, 0) + 1
    sorted_x = sorted(freq.items(), key=operator.itemgetter(1))
    sorted_dict = collections.OrderedDict(sorted_x)
    k=(list(sorted_dict.values())[-1])
    
    final_list=[]
    for ky,val in sorted_dict.items(): 
        if sorted_dict[ky]==k:
            final_list.append(tag_dic_pref[ky])
    len_of_finlist=len(final_list)
    final_list.sort()
    
    if len(final_list)==1:
        return tag_dic_pref_inverse[final_list[0]]
    if len(final_list)==2:
        return tag_dic_pref_inverse[final_list[1]]
    if len(final_list)>2:
        return tag_dic_pref_inverse[final_list[-2]]
    
```

```python
fin_data['degree_tag']=fin_data['d_tag'].apply(lambda x:ret_final_tag(x))
```
