```python
import pandas as pd
```

```python
data=pd.read_csv('/home/rk/orgs_final1.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.drop(columns=['label'],inplace=True)
```

```python
data['last_funding_status'].unique()
```

```python
def ret_fundNonFund(x):
    if x=='No_funding':
        return 0
    else:
        return 1
```

```python
data['funding_status'] = data['last_funding_status'].apply(lambda x: ret_fundNonFund(x))
```

```python
data.drop(columns=['last_funding_status'],inplace=True)
```

```python
for column in ['employee_count', 'location','industry_vertical']:
    dummies = pd.get_dummies(data[column],prefix=column)
    data[dummies.columns] = dummies
data.head()
```

```python
data.drop(columns=['employee_count', 'location','industry_vertical','company_name','description'],inplace=True)
data.head()
```

```python
data_temp=data
```

```python
unseuc_dataset=data_temp.loc[data_temp['funding_status']==0]
unseuc_dataset.shape
```

```python
succ_dataset=data_temp.loc[data_temp['funding_status']==1]
succ_dataset.shape
```

```python
tot_1=data_temp['funding_status'].value_counts()[1]
tot_0=data_temp['funding_status'].value_counts()[0]
print(tot_0,tot_1)
```

```python
data_temp.sort_values(by=['funding_status']).head()
```

```python
percent_of0_data=tot_0/(tot_1+tot_0)
num1=percent_of0_data*len(unseuc_dataset)
num2=(1-percent_of0_data)*len(unseuc_dataset)
#print(percent_of0_data*len(unseuc_dataset))
x_train_0=unseuc_dataset[:int(percent_of0_data*len(unseuc_dataset))]
x_train_1=succ_dataset[:int((percent_of0_data)*len(succ_dataset))]
x_train=pd.concat([x_train_0,x_train_1])
#print(x_train.shape)
x_test_0=unseuc_dataset[int(percent_of0_data*len(unseuc_dataset)):]
x_test_1=succ_dataset[int((percent_of0_data)*len(succ_dataset)):]
x_test=pd.concat([x_test_0,x_test_1])
#print(x_test.shape)
print(x_train.shape,x_test.shape)
```

```python
from sklearn.utils import shuffle
x_train=shuffle(x_train)
x_test=shuffle(x_test)
```

```python
y_train=x_train['funding_status']
y_train.head()
```

```python
y_test=x_test['funding_status']
y_test.head()
```

```python
x_train=x_train.drop(columns=['funding_status','no_of_funding_rounds','amount_raised','no_of_investors'])
x_test=x_test.drop(columns=['funding_status','no_of_funding_rounds','amount_raised','no_of_investors'])
```

```python
print(y_train.shape)
print(x_train.shape,x_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn import preprocessing, metrics
#from xgboost import XGBClassifier
```

```python
clfs = {
'LogisticRegression' : LogisticRegression(),
'GaussianNB': GaussianNB(),
'RandomForest': RandomForestClassifier(),
'DecisionTreeClassifier': DecisionTreeClassifier(),
'SVM': SVC(),
'KNeighborsClassifier': KNeighborsClassifier(),
'GradientBoosting': GradientBoostingClassifier(),
}
```

```python
models_report = pd.DataFrame(columns = ['Model', 'Precision_score', 'Recall_score','F1_score', 'Accuracy'])

for clf, clf_name in zip(clfs.values(), clfs.keys()):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(y_pred)
    y_score = clf.score(x_test,y_test)
    
    #print('Calculating {}'.format(clf_name))
    t = pd.Series({ 
                     'Model': clf_name,
                     'Precision_score': metrics.precision_score(y_test, y_pred),
                     'Recall_score': metrics.recall_score(y_test, y_pred),
                     'F1_score': metrics.f1_score(y_test, y_pred),
                     'Accuracy': metrics.accuracy_score(y_test, y_pred)}
                   )

    models_report = models_report.append(t, ignore_index = True)

models_report
```
